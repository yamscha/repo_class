```python
import pandas as pd
```

```python
#creating DataFrame1

raw_data_frames={"customer_Id":[10,20,40,55,78,90,33],
                 "item":["soda","chips","ice-cream","milk","onions","bread","eggs"],
                 "cost":[3.00,2.50,5.00,6.59,3.00,5.25,4.50]}
items_pd=pd.DataFrame(raw_data_frames,columns=['customer_Id',"item","cost"])
```

```python
items_pd
```

```python
#"customer_Id":[10,20,40,55,456,78,294],

raw_data_info={"name":["Krishna","Rama","Sri Ranganata","Srinivasa","Devi","Parvathi","Ganga"],
               "email":["krishna@gmail.com","rama.gmail.com","sri@hotmail.com","vasa@gmail.com",
                        "devi@hotmail.com","parvathi@yahoo.com","ganga@gmail.com"],
                "customer_Id":[10,20,40,55,456,900,333]}
data_pd = pd.DataFrame(raw_data_info,columns=['customer_Id','name','email'])

data_pd
```

```python
pd.concat([items_pd,data_pd],axis=1,verify_integrity=False)
```

```python
merge_dataframes=pd.merge(items_pd,data_pd,on="customer_Id")
merge_dataframes
```

```python
merge_tables=pd.merge(items_pd,data_pd,on="customer_Id",how='right')
merge_tables
```

```python
merge_tables=pd.merge(items_pd,data_pd,on="customer_Id",how='outer')
merge_tables
```

```python
merge_tables=pd.merge(items_pd,data_pd,on="customer_Id",how='right')
merge_tables
```
