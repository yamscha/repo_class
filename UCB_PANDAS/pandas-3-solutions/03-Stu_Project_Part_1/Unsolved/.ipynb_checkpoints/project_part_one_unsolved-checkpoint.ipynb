```python
# Dependencies
import pandas as pd
```

```python
# load CSV

new_coders = "../Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas

new_coders_pd= pd.read_csv(new_coders, encoding="iso-8859-1", low_memory=False)
new_coders_pd.head()
```

```python
# Inspect all columns

new_coders_pd.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111 

reduced_coders_pd = new_coders_pd.iloc[:,[0,1,2,3,4,7,8,9,10,11,29,30,32,36,37,45,48,56,110,111]]

reduced_coders_pd.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
reduced_coders_pd = reduced_coders_pd.replace({0:"No",1:"Yes"})
reduced_coders_pd.head()
```

```python
# Calculate total number of respondents in survey

total_surveyed=len(reduced_coders_pd)
total_surveyed
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"] == "Yes"]
attended_bootcamp.head()
```

```python
# Calculate average age of attendees 
average_age = attended_bootcamp["Age"].mean()
print(average_age)
```

```python
# Calculate how many people attended a bootcamp
count_people=attended_bootcamp["AttendedBootcamp"].count()
print(count_people)
```

```python
#Calculate how many attendees hold degrees
count_hold_degrees=attended_bootcamp["SchoolDegree"].count()
#count_hold_degrees=attended_bootcamp["SchoolDegree"].value_counts()
print(count_hold_degrees)
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification  
total_no_attendees = attended_bootcamp["Gender"].count()
male_count = attended_bootcamp["Gender"].value_counts()["male"]
female_count = attended_bootcamp["Gender"].value_counts()["female"]
print("Total:",total_no_attendees)
print("Male_Count:",male_count)
print("Female_Count:",female_count)
```

```python
# Calculate percentage of respondents who attended a bootcamp
percentage_respondents =count_people/total_surveyed*100
print(percentage_respondents)
```

```python
# Calculate percentage of respondents belonging to each gender
male_gender_percent = male_count/total_no_attendees*100
female_gender_percent = female_count/total_no_attendees*100

print("male_gender_percent:",male_gender_percent)
print("female_gender_percent:",female_gender_percent)
```

```python
# Calculate percentage of attendees with a school degree
percentage_school_degree=count_hold_degrees/count_people*100

print(percentage_school_degree)
```

```python
# Calculate average post-bootcamp salary
average_post_bootcamp_salary=attended_bootcamp["BootcampPostSalary"].mean()
print(average_post_bootcamp_salary)
```
