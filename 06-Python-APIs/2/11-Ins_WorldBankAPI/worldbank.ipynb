```python
# Dependencies
import requests
```

```python
url = "http://api.worldbank.org/"
file_format = "json"
full_url = f"{url}countries?format={file_format}"
full_url
```

```python
# Get country information in JSON format
countries_response = requests.get(full_url).json()
```
