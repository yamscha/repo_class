```python
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
api_key = "164b73c522a8420c8e05343ef1da0a7e"
```

```python
def format_query(url,api_key,q):
    return f"{url}api-key={api_key}&q={q}"
```

```python
# Search for articles that mention Russian Bot
q = "russian bot"
```

```python
# Build query URL
query = format_query(url,api_key,q)
query
```

```python
# Populate articles
articles = requests.get(query).json()
```

```python
articles
```

```python
article_l = articles["response"]["docs"]
article_l
```

```python
# The "response" property in articles contains the actual articles
# list comprehension.
url_l = [article["web_url"] for article in article_l]
```
