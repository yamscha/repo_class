```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
csvpath = os.path.join('Resources', 'avg_rain_state.csv')
rain_df = pd.read_csv(csvpath)
rain_df.head()
```

```python
# Set x values
x_values = np.arange(len(rain_df))
```

```python
# Set figure size to adjust for space
plt.figure(figsize=(20, 5))
# Create a bar chart
plt.bar(x_values, rain_df["Inches"], color='r', alpha=0.5)
```

```python
# Create ticks
#add a semicolon if you don't want to see all the output
plt.xticks(x_values, rain_df["State"], rotation="vertical"); 
```

```python
# Set x and y limits
plt.xlim(-1, len(x_values))
```

```python
plt.ylim(0, max(rain_df["Inches"])+10)
```

```python
# Set a Title and labels
plt.title("Average Rain per State")
plt.xlabel("State")
plt.ylabel("Average Amount of Rainfall in Inches")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("avg_state_rain.png")
plt.show()
```

### Using Pandas to Chart a DataFrame

```python
# Filter the DataFrame down only to those columns to chart
state_and_inches = rain_df[["State","Inches"]]

# Set the index to be "State" so they will be used as labels
state_and_inches = state_and_inches.set_index("State")

state_and_inches.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart
# of the data
state_and_inches.plot(kind="bar", 
                      figsize=(20, 5),
                      title="Average Rain Per State")

# Pandas uses matplotlib, so you can still edit the chart
# in the same way you would edit a matplotlib bar chart.

# Set a title for the chart
plt.title("Average Rain Per State!!!")

plt.show()
```

```python
# Pandas can also plot multiple columns if the DataFrame
# includes them
multi_plot = rain_df.plot(kind="bar", figsize=(20, 5))
print(type(multi_plot))

# PandasPlot.set_xticklabels() can be used
# to set the tick labels as well
multi_plot.set_xticklabels(rain_df["State"], rotation=45)

plt.show()
```
