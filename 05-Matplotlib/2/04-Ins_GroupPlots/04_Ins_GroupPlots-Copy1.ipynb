```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Import our data into pandas from CSV
csvpath = os.path.join('Data', 'UsedCars.csv')
used_car_df = pd.read_csv(csvpath)

used_car_df.head()
```

```python
"""
If we don't select a single column, notice that it counts all columns

"""
# Create a group based on the values in the 'maker' column
maker_group = used_car_df.groupby('maker')
# Count how many times each maker appears in our group
count_makers = maker_group.count()

count_makers
```

```python
# Create a group based on the values in the 'maker' column
maker_group = used_car_df.groupby('maker')
# Count how many times each maker appears in our group
count_makers = maker_group['manufacture_year'].count()

count_makers
```

```python
"""
Notice that we can start with the dataframe nad call 
.value_counts (no groupby).
"""
#this function is pretty similar to .counts()
used_car_df['maker'].value_counts() 
```

```python
# Create a pie chart based off of the group series from before
count_chart = count_makers.plot(kind='bar', figsize=(9, 9))
print(type(count_chart))
```

```python
# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Car Manufacturer")
count_chart.set_ylabel("Number of Cars")
# plt.xlabel("Car Manufacturer")
# plt.ylabel("Number of Cars")

plt.show()
```
