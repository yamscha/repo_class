```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club",
        "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_values = np.arange(len(gyms))

plt.title("NYC Gym Popularity")
plt.xlabel("Gym Name")
plt.ylabel("Number of Members")

plt.bar(x_values, members, facecolor="red",
        alpha=0.75, tick_label=gyms)
plt.show()
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_values = np.arange(0, x_lim, 0.1)
sin = np.sin(x_values)
```

```python
plt.title("Sin from 0 to 2$\pi$")
plt.xlabel("Real Numbers from 0 to 2$\pi$")
plt.ylabel("sin(x)")

plt.hlines(0, 0, x_lim, alpha=0.2)
plt.xlim(0, x_lim)
plt.ylim(-1.25, 1.25)

plt.plot(x_values, sin, marker="o", color="red", linewidth=1)
plt.show()
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club",
        "Rickie's Gym"]
members = [49, 92, 84, 53]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = [0, 0.05, 0, 0]
```

```python
plt.title("NYC Gym Popularity")
plt.pie(members, explode=explode, labels=gyms, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```

```python
# DATASET 4
x_values = np.arange(0, 10, 0.1)
times = []
max_x = np.ceil(max(x_values))
for x in x_values:
    times.append(x * x + np.random.randint(0, max_x))
# times = [(x*x+np.random.randint(0, max_x)) for x in x_values]
```

```python
plt.title("Running Time of FakeSort for Sample Input Sizes")
plt.xlabel("Length of Input Array")
plt.ylabel("Time to Sort (s)")

plt.scatter(x_values, times, marker="o", color="red")
plt.show()
```
